```python
# 선형!------------------------------------------------------------------------------------
xData=[1,2,3,4,5,6,7]
yData=[25,55,75,110,128,155,180]
W=tf.Variable(tf.random_uniform([1],-100,100))
b=tf.Variable(tf.random_uniform([1],-100,100))
X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
H=W*X+b
```

```python
cost=tf.reduce_mean(tf.square(H-Y))
a=tf.Variable(0.01)
optimizer=tf.train.GradientDescentOptimizer(a)
train=optimizer.minimize(cost)
```

```python
init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
for i in range(5001):
    sess.run(train,feed_dict={X:xData,Y: yData})
    if i%500 ==0:
        print(i, sess.run(cost,feed_dict={X:xData,Y:yData}),sess.run(W),sess.run(b))
        
print(sess.run(H,feed_dict={X:[8]}))
```

```python
#---------------------------------------폴더내에 pdata로 학습 시키기
import tensorflow.compat.v1 as tf
import numpy as np
from pandas.io.parsers import read_csv

model=tf.global_variables_initializer()
data=read_csv('pdata.csv',sep=',')   # , 으로 분류하기 
xy=np.array(data,dtype=np.float32)
print(xy)
```

```python
#pdata 의 형식은 year,avgTemp,minTemp,maxTemp,rainFall,avgPrice
#20100101,-4.9,-11,0.9,0,2123
#20100102,-3.1,-5.5,5.5,0.8,2123
#요런식....
x_data=xy[:,1:-1]
y_data=xy[:,[-1]]
```

```python
X=tf.placeholder(tf.float32,shape=[None,4])
Y=tf.placeholder(tf.float32,shape=[None,1])
W=tf.Variable(tf.random_normal([4,1]),name="weight")
b=tf.Variable(tf.random_normal([1]),name="bias")
```

```python
hypothesis=tf.matmul(X,W)+b
cost=tf.reduce_mean(tf.square(hypothesis-Y))
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.000005)
train=optimizer.minimize(cost)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(100001):

    cost_, hypo_, _ = sess.run([cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})

    if step % 500 == 0:

        print("#", step, " 손실 비용: ", cost_)

        print("- 배추 가격: ", hypo_[0])
```

```python
saver=tf.train.Saver()
save_path=saver.save(sess,"./saved.cpkt")
#     checkpoint / cpkt.data / cpkt.meta/ cpkt.index 가 생성됨
```

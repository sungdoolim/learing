```python
def calc_bmi(h,w):
    bmi=w/(h/100)**2
    if bmi<18.5: return "thin"
    if bmi<25: return "normal"
    return "fat"
fp=open("bmi.csv","w",encoding="utf-8")
fp.write("height,weight,label\r\n")
cnt={"thin":0,"normal":0,"fat":0}
for i in range(20000):
    h=random.randint(120,200)
    w=random.randint(35,80)
    label=calc_bmi(h,w)
    cnt[label]+=1
    fp.write("{0},{1},{2}\r\n".format(h,w,label))
fp.close()
#파일 만들기로 안해도 될듯.... bmi.csv생성 하는 코드임
'''
height,weight,label

190,67,normal

130,79,fat

......
....

'''
```

```python
csv=pd.read_csv("bmi.csv")
csv["weight"]/=100
csv["height"]/=200
print(csv)
```

```python
bmi_class={
    "thin":[1,0,0],
    "normal":[0,1,0],
    "fat":[0,0,1]
}

y=np.empty((20000,3))
for i,v in enumerate(csv["label"]):
    y[i]=bmi_class[v]
    print(i," ",v," ",y[i])
```

```python
x=csv[["weight","height"]].values
print(x)
```

```python
x_train,y_train,x_label,y_label=train_test_split(x,y)
print(x_train)
print(y_label)
```

```python
model=Sequential()
model.add(Dense(512,input_shape=(2,)))
model.add(Activation('relu'))
model.add(Dropout(0,1))
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0,1))
model.add(Dense(3))
model.add(Activation('softmax'))
model.compile("rmsprop","categorical_crossentropy",metrics=['accuracy'])

model.fit(x_train,x_label,batch_size=30,nb_epoch=5)
#model.save('mnist_mlp_model.h5')모델저장
score=model.evaluate(y_train,y_label)
print("loss: ",score[0])
print("accurach: ",score[1])
```

```python
#model = load_model('mnist_mlp_model.h5')#모델 불러오기

score=model.evaluate(y_train,y_label)
print("loss: ",score[0])
print("accurach: ",score[1])
```

```python
files=codecs.open("n.txt","r",encoding="utf-8")
soup=BeautifulSoup(files,"html.parser")
body=soup.select_one("body")
text=body.getText()
#print(text)
# 네이버 페이지 가져온거 실행 안하는게 좋을듯..
```

```python
twitter=Twitter()
word_dic={}
lines=text.split("\r\n")
#print(lines)
```

```python
for line in lines :
    malist=twitter.pos(line)
   # print(malist)  #[('\n\n\n', 'Foreign'), ('연합뉴스', 'Noun'), ....]
    for taeso,pumsa in malist:
        if pumsa=="Noun":
            if not (taeso in word_dic):
                word_dic[taeso]=0# 처음에 dic 만들어 내는 코드로 필수임
            word_dic[taeso]+=1
print(word_dic)
```

```python
files=codecs.open("n.txt","r",encoding="utf-8")
soup=BeautifulSoup(files,"html.parser")
body=soup.select_one("body")
text=body.getText()
twitter=Twitter()
lines=text.split("\r\n")
results=[]
print(lines)
```

```python
for line in lines:
    r=[]
    malist=twitter.pos(line,norm=True,stem=True)
    for(word,pumsa) in malist:
        if not pumsa in ["Josa","Eomi","Punctuation"]:
            r.append(word)# 한 줄이 저장됨
            #print(r)
    results.append((" ".join(r)).strip())# 처음~ 끝까지 모두 저장됨
    # 형식은 ['\n\n\n', '연합뉴스', '바로가기', '\n', '주']
output=(" ".join(results)).strip()
#print(results)
```

```python
with open ("n1.wakati","w",encoding="utf-8") as fp:
    fp.write(output)
    # 파일 만들기고....
data=word2vec.LineSentence("n1.wakati")
print(data)# 객체 형식으로 저장되 있네 이상한 값 한줄 나옴
model=word2vec.Word2Vec(data,size=400,window=10,hs=1,min_count=2,sg=1)# size는 우리가 정하는 건가바
model.save("n1.model")# 모델 저장
```
